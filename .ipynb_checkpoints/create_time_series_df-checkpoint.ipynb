```python
# generate features / technical indicators functions

## RSI
def change_val(stock_df):
    A = []
    for i in range(len(stock_df)):
        A.append(stock_df['Close'][i]-stock_df['Close'][i-1])
    A[0]=0
    return(A)

def gain(A):
    B=A.copy()
    for i in range(len(B)):
        if B[i]<0:
            B[i]=0
    return(B)

def loss(A):
    C=A.copy()
    for i in range(len(C)):
        if C[i]>0:
            C[i]=0
        else:
            C[i] = abs(C[i])
    return(C)

def RS_Val(gain,loss,n):
    D=[0]*len(gain)
    for i in range(len(D)-n):
        D[i+n-1]= sum(gain[i:i+n])/sum(loss[i:i+n])
    return(D)

def RSI(RSV):
    E=[0]*len(RSV)
    for i in range(len(E)):
        E[i]=100-100/(1+RSV[i])
    return(E)

def get_RSI(stock_df,n):
    A = change_val(stock_df)
    B = gain(A)
    C = loss(A)
    RSV = RS_Val(B,C,n)
    return(RSI(RSV))

##SMA
def get_SMA(stock_df,n):
    A = [0]*len(stock_df)
    for i in range(len(stock_df)-n):
        A[i+n-1] = np.mean(list(stock_df['Close'][i:i+n]))
    return(A)

#EMA
def get_EMA(stock_df,n):
    mult = 2/(n+1)
    A = [0]*len(stock_df)
    A[n-1] = get_SMA(stock_df,n)[n-1]
    for i in range(n-1,len(stock_df)-1):
        A[i+1] = A[i]+(stock_df['Close'][i]-A[i])*mult
    return(A)

#vol change
def change_vol_pct(stock_df):
    A = [0]*len(stock_df)
    for i in range(len(stock_df)-1):
        A[i+1] = (stock_df['Volume'][i+1]-stock_df['Volume'][i])/stock_df['Volume'][i]
    return(A)

#pct change
def val_change_pct(stock_df):
    A = [0]*len(stock_df)
    for i in range(len(stock_df)-1):
        A[i+1] = (stock_df['Close'][i+1]-stock_df['Close'][i])/stock_df['Close'][i]
    return(A)

def SMA_next(stock_df,n):
    A=[0]*len(stock_df)
    for i in range(len(stock_df)-n):
        A[i] = np.mean(stock_df['Close'][i:i+n])
    return(A)

def decision(stock_df):
    A[i] = [0]*len(stock_df)
    for i in range(len(A)):
        if stock_df['pct change price']>0:
            A[i]=1
    return(A)
```

```python
SPY = web.DataReader('SPY',source,start, end)

def consol_df(stock_df):
    n1 = 9
    n2 = 12
    n3 = 26
    n4 = 14
    stock_df['RSI'] = get_RSI(stock_df,n1)
    stock_df['pct change volume'] = change_vol_pct(stock_df)
    stock_df['pct change price'] = val_change_pct(stock_df)
    stock_df['12 EMA'] = get_EMA(stock_df,n2)
    stock_df['26 EMA'] = get_EMA(stock_df,n3)
    stock_df['14 F SMA'] = SMA_next(stock_df,n4)
    return(stock_df)[n3-1:len(stock_df)-n4]
consol_df(SPY)
```
